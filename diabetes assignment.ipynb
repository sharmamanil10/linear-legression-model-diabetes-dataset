```python
import numpy as np
data=np.genfromtxt("training_diabetes_x_y_train.csv",delimiter=",")
testcsv=np.genfromtxt("test_diabetes_x_test.csv",delimiter=",")
x=data[:,0:10]
y=data[:,10]
x_test=testcsv[:,0:10]

```

```python

from sklearn import model_selection

from sklearn.linear_model import LinearRegression
alg1=LinearRegression()
alg1.fit(x,y)
y_pred=alg1.predict(x_test)
np.savetxt("y-prediction.csv",y_pred,delimiter=",")
import matplotlib.pyplot as plt

x=np.arange(0,350,0.1)
plt.plot(x,x)
plt.show()
y_pred.shape
```
